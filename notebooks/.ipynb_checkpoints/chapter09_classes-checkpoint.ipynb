```python
# åŸºæœ¬ç±»å®šä¹‰æ¼”ç¤º
print("=== åŸºæœ¬ç±»å®šä¹‰æ¼”ç¤º ===")

class BankAccount:
    """é“¶è¡Œè´¦æˆ·ç±»"""
    
    def __init__(self, account_number, owner_name, initial_balance=0):
        """åˆå§‹åŒ–è´¦æˆ·"""
        self.account_number = account_number
        self.owner_name = owner_name
        self.balance = initial_balance
    
    def deposit(self, amount):
        """å­˜æ¬¾"""
        if amount > 0:
            self.balance += amount
            return f"å­˜å…¥ {amount}ï¼Œå½“å‰ä½™é¢: {self.balance}"
        return "å­˜æ¬¾é‡‘é¢å¿…é¡»å¤§äº0"
    
    def withdraw(self, amount):
        """å–æ¬¾"""
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"å–å‡º {amount}ï¼Œå½“å‰ä½™é¢: {self.balance}"
        return "ä½™é¢ä¸è¶³æˆ–é‡‘é¢æ— æ•ˆ"
    
    def get_balance(self):
        """æŸ¥è¯¢ä½™é¢"""
        return self.balance
    
    def __str__(self):
        """å­—ç¬¦ä¸²è¡¨ç¤º"""
        return f"è´¦æˆ·({self.account_number}): {self.owner_name}, ä½™é¢: {self.balance}"

# åˆ›å»ºå’Œä½¿ç”¨ç±»å®ä¾‹
account = BankAccount("12345", "å¼ ä¸‰", 1000)
print(f"åˆ›å»ºè´¦æˆ·: {account}")

print(account.deposit(500))
print(account.withdraw(200))
print(f"å½“å‰ä½™é¢: {account.get_balance()}")

# è®¿é—®å®ä¾‹å±æ€§
print(f"è´¦æˆ·æ‰€æœ‰è€…: {account.owner_name}")
print(f"è´¦æˆ·å·ç : {account.account_number}")
```

```python
# ç»§æ‰¿æ¼”ç¤º
print("\=== ç»§æ‰¿æ¼”ç¤º ===")

class SavingsAccount(BankAccount):
    """å‚¨è“„è´¦æˆ·ç±»ï¼Œç»§æ‰¿è‡ªBankAccount"""
    
    def __init__(self, account_number, owner_name, initial_balance=0, interest_rate=0.02):
        """åˆå§‹åŒ–å‚¨è“„è´¦æˆ·"""
        super().__init__(account_number, owner_name, initial_balance)
        self.interest_rate = interest_rate
    
    def add_interest(self):
        """æ·»åŠ åˆ©æ¯"""
        interest = self.balance * self.interest_rate
        self.balance += interest
        return f"æ·»åŠ åˆ©æ¯ {interest:.2f}ï¼Œå½“å‰ä½™é¢: {self.balance:.2f}"
    
    def __str__(self):
        """é‡å†™å­—ç¬¦ä¸²è¡¨ç¤º"""
        return f"å‚¨è“„è´¦æˆ·({self.account_number}): {self.owner_name}, ä½™é¢: {self.balance:.2f}, åˆ©ç‡: {self.interest_rate*100}%"

# åˆ›å»ºå‚¨è“„è´¦æˆ·
savings = SavingsAccount("54321", "æå››", 2000, 0.03)
print(f"åˆ›å»ºå‚¨è“„è´¦æˆ·: {savings}")

# ä½¿ç”¨ç»§æ‰¿çš„æ–¹æ³•
print(savings.deposit(500))

# ä½¿ç”¨å­ç±»ç‰¹æœ‰çš„æ–¹æ³•
print(savings.add_interest())

print(f"æœ€ç»ˆçŠ¶æ€: {savings}")

# æ£€æŸ¥ç»§æ‰¿å…³ç³»
print(f"\isinstance(savings, SavingsAccount): {isinstance(savings, SavingsAccount)}")
print(f"isinstance(savings, BankAccount): {isinstance(savings, BankAccount)}")
print(f"issubclass(SavingsAccount, BankAccount): {issubclass(SavingsAccount, BankAccount)}")
```

```python
# è¿­ä»£å™¨æ¼”ç¤º

print("=== è¿­ä»£å™¨æ¼”ç¤º ===")


class Counter:
    
    """è®¡æ•°å™¨è¿­ä»£å™¨"""        
    def __init__(self, max_count):        
        self.max_count = max_count        
        self.current = 0        
    def __iter__(self):        
        """è¿”å›è¿­ä»£å™¨å¯¹è±¡æœ¬èº«"""        
        return self        
    def __next__(self):        
        """è¿”å›ä¸‹ä¸€ä¸ªå€¼"""        
        if self.current < self.max_count:            
            self.current += 1            
            return self.current        
        else:            
            raise StopIteration # ä½¿ç”¨è¿­ä»£å™¨
            
print("è®¡æ•°å™¨è¿­ä»£å™¨:")

counter = Counter(5)

for num in counter:    
    print(f"  è®¡æ•°: {num}")
    
# æ‰‹åŠ¨è¿­ä»£

print("\æ‰‹åŠ¨è¿­ä»£:")
counter2 = Counter(3)
iterator = iter(counter2)
try:    
    while True:        
        value = next(iterator)        
        print(f"  æ‰‹åŠ¨è·å–: {value}")
except StopIteration:    
    print("  è¿­ä»£å®Œæˆ")
    # ç”Ÿæˆå™¨å‡½æ•°
    
def fibonacci_generator(n):    
    """æ–æ³¢é‚£å¥‘ç”Ÿæˆå™¨"""    
    a, b = 0, 1    
    count = 0    
    while count < n:        
        yield a        
        a, b = b, a + b        
        count += 1
        print("\æ–æ³¢é‚£å¥‘ç”Ÿæˆå™¨:")
        
for i, fib in enumerate(fibonacci_generator(8)):    
    print(f"  F({i}) = {fib}") 
    
# ç”Ÿæˆå™¨è¡¨è¾¾å¼
print("ç”Ÿæˆå™¨è¡¨è¾¾å¼:")

squares_gen = (x**2 for x in range(5))

print(f"ç”Ÿæˆå™¨ç±»å‹: {type(squares_gen)}")

for square in squares_gen:    
    print(f"  å¹³æ–¹: {square}")
    
# ç´ æ•°ç”Ÿæˆå™¨
def primes_generator(max_num):    
    """ç´ æ•°ç”Ÿæˆå™¨"""    
    def is_prime(n):        
        if n < 2:            
            return False        
        for i in range(2, int(n ** 0.5) + 1):            
            if n % i == 0:                
                return False        
        return True        
            
    for num in range(2, max_num + 1):        
        if is_prime(num):            
            yield num
        
print("ç´ æ•°ç”Ÿæˆå™¨:")

primes = list(primes_generator(20))
print(f"20ä»¥å†…çš„ç´ æ•°: {primes}")
```
