```python
# äºŒè¿›åˆ¶æ•°æ®å¤„ç†æ¼”ç¤º
print("\
=== äºŒè¿›åˆ¶æ•°æ®å¤„ç†æ¼”ç¤º ===")

import struct

# æ‰“åŒ…æ•°æ®
data = struct.pack('i4sf', 123, b'spam', 3.14)
print(f"æ‰“åŒ…åçš„äºŒè¿›åˆ¶æ•°æ®: {data}")
print(f"æ•°æ®é•¿åº¦: {len(data)} å­—èŠ‚")

# è§£åŒ…æ•°æ®
unpacked = struct.unpack('i4sf', data)
print(f"è§£åŒ…åçš„æ•°æ®: {unpacked}")

# å¤„ç†ä¸åŒçš„æ•°æ®ç±»å‹
print("\
ä¸åŒæ•°æ®ç±»å‹ç¤ºä¾‹:")
formats = [
    ('æ•´æ•°', 'i', 42),
    ('æµ®ç‚¹æ•°', 'f', 3.14),
    ('åŒç²¾åº¦', 'd', 2.718281828),
    ('å­—ç¬¦', 'c', b'A'),
    ('å­—ç¬¦ä¸²', '10s', b'hello')
]

for name, fmt, value in formats:
    packed = struct.pack(fmt, value)
    unpacked = struct.unpack(fmt, packed)[0]
    print(f"  {name}: {value} -> {packed} -> {unpacked}")

# å­—èŠ‚åºå¤„ç†
print("\
å­—èŠ‚åºæ¼”ç¤º:")
value = 0x12345678

# å¤§ç«¯åº
big_endian = struct.pack('>I', value)
print(f"å¤§ç«¯åº: {big_endian.hex()}")

# å°ç«¯åº
little_endian = struct.pack('<I', value)
print(f"å°ç«¯åº: {little_endian.hex()}")

# ç³»ç»Ÿå­—èŠ‚åº
native = struct.pack('I', value)
print(f"ç³»ç»Ÿå­—èŠ‚åº: {native.hex()}")"
```

```python
# å¤šçº¿ç¨‹æ¼”ç¤º
print("\
=== å¤šçº¿ç¨‹æ¼”ç¤º ===")

import threading
import time
import queue

# ç®€å•çš„çº¿ç¨‹å‡½æ•°
def worker(name, delay):
    """å·¥ä½œçº¿ç¨‹å‡½æ•°"""
    print(f"çº¿ç¨‹ {name} å¼€å§‹å·¥ä½œ")
    time.sleep(delay)
    print(f"çº¿ç¨‹ {name} å®Œæˆå·¥ä½œ")

# åˆ›å»ºå’Œå¯åŠ¨çº¿ç¨‹
print("åˆ›å»ºå¤šä¸ªå·¥ä½œçº¿ç¨‹:")
threads = []
for i in range(3):
    t = threading.Thread(target=worker, args=(f"Worker-{i+1}", 0.1))
    threads.append(t)
    t.start()

# ç­‰å¾…æ‰€æœ‰çº¿ç¨‹å®Œæˆ
for t in threads:
    t.join()

print("æ‰€æœ‰å·¥ä½œçº¿ç¨‹å®Œæˆ")

# ä½¿ç”¨é”è¿›è¡ŒåŒæ­¥
print("\
=== çº¿ç¨‹åŒæ­¥æ¼”ç¤º ===")

shared_counter = 0
lock = threading.Lock()

def increment_counter(name, count):
    """å¢åŠ è®¡æ•°å™¨"""
    global shared_counter
    for i in range(count):
        with lock:  # ä½¿ç”¨é”ä¿æŠ¤å…±äº«èµ„æº
            current = shared_counter
            time.sleep(0.001)  # æ¨¡æ‹Ÿå¤„ç†æ—¶é—´
            shared_counter = current + 1
    print(f"{name} å®Œæˆ {count} æ¬¡å¢é‡")

# é‡ç½®è®¡æ•°å™¨
shared_counter = 0

# åˆ›å»ºå¤šä¸ªçº¿ç¨‹åŒæ—¶ä¿®æ”¹å…±äº«å˜é‡
sync_threads = []
for i in range(3):
    t = threading.Thread(target=increment_counter, args=(f"Thread-{i+1}", 10))
    sync_threads.append(t)
    t.start()

for t in sync_threads:
    t.join()

print(f"æœ€ç»ˆè®¡æ•°å™¨å€¼: {shared_counter}")

# ç”Ÿäº§è€…-æ¶ˆè´¹è€…æ¨¡å¼
print("\
=== ç”Ÿäº§è€…-æ¶ˆè´¹è€…æ¨¡å¼ ===")

data_queue = queue.Queue(maxsize=5)

def producer(name, items):
    """ç”Ÿäº§è€…"""
    for i in range(items):
        item = f"{name}-item-{i+1}"
        data_queue.put(item)
        print(f"ç”Ÿäº§è€… {name} ç”Ÿäº§: {item}")
        time.sleep(0.1)
    print(f"ç”Ÿäº§è€… {name} å®Œæˆ")

def consumer(name, items):
    """æ¶ˆè´¹è€…"""
    for i in range(items):
        item = data_queue.get()
        print(f"æ¶ˆè´¹è€… {name} æ¶ˆè´¹: {item}")
        data_queue.task_done()
        time.sleep(0.15)
    print(f"æ¶ˆè´¹è€… {name} å®Œæˆ")

# å¯åŠ¨ç”Ÿäº§è€…å’Œæ¶ˆè´¹è€…
producer_thread = threading.Thread(target=producer, args=("Producer-1", 5))
consumer_thread = threading.Thread(target=consumer, args=("Consumer-1", 5))

producer_thread.start()
consumer_thread.start()

producer_thread.join()
consumer_thread.join()

print("ç”Ÿäº§è€…-æ¶ˆè´¹è€…æ¼”ç¤ºå®Œæˆ")"
```
